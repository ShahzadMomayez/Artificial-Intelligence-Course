# Data preprocessing
- write BrainDataset class
-resize all to 512* 512

```python
class BrainDataset(Dataset):
    def __init__(self):
        self.root_dir=rootDir
        self.classes =  classes
        self.samples =[]
        self.labels =[]

        transform = transforms.Compose([
            transforms.Resize((img_size, img_size)),
            transforms.ToTensor(),
         #   transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Example values for normalization
        ])

        for i, cls in enumerate(self.classes):
            cls_folder = os.path.join(self.root_dir, cls)
            for img_file in os.listdir(cls_folder):
                img_path = os.path.join(cls_folder, img_file)
                image = Image.open(img_path).convert('RGB')

                image_tensor = transform(image)
                self.samples.append(image_tensor)
                self.labels.append(i)

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        return self.samples[idx], self.labels[idx]


data_set =BrainDataset()
```

# Normalization
Normalization is a technique often applied in data preprocessing to adjust the values of an input dataset to a common scale without distorting differences in the ranges of values or losing information. It is commonly used in the context of machine learning and deep learning, as well as in other fields such as statistics and data analysis.

Normalization can be crucial for the performance of many machine learning algorithms, especially optimization algorithms used in neural networks, like gradient descent. It’s also important for image processing and is routinely applied to neural network inputs and datasets.

In the context of deep learning, normalization might refer to several different processes:

-  `Feature Scaling`: In this process, the data features are scaled so that they have the properties of a standard normal distribution with μ = 0 and σ = 1, where μ is the mean (average) and σ is the standard deviation from the mean. Standard scores (also called z-scores) of the samples are calculated as follows:

$${z = (x - μ) / σ}$$

- `Min-Max Normalization`: This technique scales the data to a fixed range - usually 0 to 1. The general formula for a min-max normalization is:

$${X_norm = (X - X_min) / (X_max - X_min) }$$

Where X_max and X_min are the maximum and the minimum values of the feature respectively.

- `Batch Normalization`: In the context of neural networks, batch normalization is a technique to standardize the inputs to a layer for each mini-batch. This stabilizes the learning process and dramatically reduces the number of training epochs required to train deep networks.

 normalization changes the range and distribution of the pixel intensity values.

## Advantages of Normalization:

- Improved Model Training:
Normalization helps in faster convergence during neural network training. Without normalization, individual features could vary in range and scale, causing the training process to be longer and less stable.

- Equal Contribution:
For machine learning algorithms, particularly those that compute distances between data points (like k-nearest neighbors), normalization ensures that each feature contributes equally to the distance computations.

- Better Gradient Descent Behavior:
With features on the same scale, the gradients for model weights update more uniformly, avoiding oscillations or slow learning in the weight space.

- Prevents Saturation:
In deep learning, normalization can help prevent neurons from saturating in regions where activation functions yield almost zero gradients, which can stop the model from further learning.

- Stabilizes Training:
Techniques like batch normalization can stabilize neural networks by maintaining the mean and variance of the layers’ inputs constant throughout the training.

- Better Generalization
.

- Controls Overfitting

## Disadvantagous

- performance impact
- complexity


### there are certain situations where normalization might not be beneficial or necessary:

1- Data is Already in a Consistent Scale: If your data features are already on a similar scale or have the same range, normalization might not be needed. This is often the case with image data, where pixel values are usually in the range of 0-255.

2- Homogeneous Data: When the data is homogeneous (i.e., all the features are of the same type and similar scales), normalization may not provide any significant benefit.

3- Tree-Based Models

# Image classification using CNN network

```python
batch_size = 4
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=4, shuffle=False)
class CNNModel(nn.Module):
    def __init__(self, num_classes):
        super(CNNModel, self).__init__()
        self.conv_ = nn.Conv2d(3, 16, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.dropout = nn.Dropout(0.5)
        self.linear = nn.Linear(16 * 256 * 256, num_classes)

    def forward(self, x):
        x = self.pool(F.relu(self.conv_(x)))
        x = x.view(-1, 16 * 256 * 256)
        x = self.dropout(x)
        x = self.linear(x)
        return x



```

```python
model = CNNModel(len(classes))
loss_func = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
num_epochs = 8

train_losses = []
val_accurs = []

def train_an_epoc(model, train_loader, optimizer, loss_func):
    model.train()
    total_loss = 0.0

    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = loss_func(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss =total_loss+ loss.item()
    epoc_loss = total_loss / len(train_loader)
    train_losses.append(epoc_loss)
    return epoc_loss

def measure_accuracy(model, test_loader):
    model.eval()
    true_val = 0
    total = 0

    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            true_val += (predicted == labels).sum().item()
    epoc_accur = 100 * true_val / total
    val_accurs.append(epoc_accur)
    return epoc_accur

for epoch in range(num_epochs):
    train_loss = train_an_epoc(model, train_loader, optimizer, loss_func)
    print(f"epoch {epoch+1}")
    accuracy = measure_accuracy(model, test_loader)
    print(f"accuracy: {accuracy}%")
    print("-----------------------")
```

# Cost Function

### Mean Squared Error (MSE):

- How it works: The MSE measures the average squared difference between the estimated values and the actual value. It is calculated as
$${ MSE = (1/n) ∑(actual - predicted)^2}$$

- Application: Regression problems.
- Reason for choosing: Simple to understand and implement, differentiable, and works well in many scenarios where the errors are normally distributed.
### Mean Absolute Error (MAE):

- How it works: MAE measures the average magnitude of errors in a set of predictions, without considering their directions. It is computed as:
$${ MAE = (1/n) ∑actual - predicted}$$

- Application: Regression problems.
- Reason for choosing: Less sensitive to outliers compared to MSE.
### Cross-Entropy Loss (or Log Loss):

- How it works: Measures the performance of a classification model whose output is a probability between 0 and 1. Cross-entropy loss increases as the predicted probability diverges from the actual label.
- Application: Binary and multi-class classification problems.
- Reason for choosing: When dealing with probabilities, it handles extreme predictions better than Mean Squared Error, and it inherently takes the uncertainty of the prediction into account.
### Hinge Loss:

- How it works: It is primarily used with Support Vector Machine (SVM) models. It is calculated as
$${ Hinge Loss = max(0, 1 - (actual * predicted)) }$$ for binary classification.
- Application: Binary classification, especially for support vector machines.
- Reason for choosing: Encourages the model to not just correctly classify the data but also to maximize the margin between the classes.

# Optimizer --> Adam Optimizer

The Adam (Adaptive Moment Estimation) optimizer is an adaptive learning rate optimization algorithm designed specifically for training deep neural networks. A Method for Stochastic Optimization. It is considered one of the most effective optimization algorithms for training neural models because of its capability of combining the advantages of two other extensions of stochastic gradient descent, namely:

- Adaptive Gradient Algorithm (AdaGrad) that maintains a per-parameter learning rate that improves performance on problems with sparse gradients (like natural language and computer vision problems).

- Root Mean Square Propagation (RMSProp) that also maintains per-parameter learning rates that are adapted based on the average of recent magnitudes of the gradients for the weight (which means it works well on online and non-stationary problems).

### Here’s how Adam works:

Adam computes adaptive learning rates for each parameter by estimating the first (the mean) and the second (the uncentered variance) moments of the gradients.
Unlike SGD which maintains a single learning rate for all weight updates and the learning rate does not change during training, Adam’s learning rate is calculated from the estimation of first and second moments of the gradients (hence adaptive).
Adam’s stepsizes are approximately bounded by the stepsize hyperparameter, making it somewhat invariant to the scale of the gradients.
Adam includes bias corrections to the first and second moment estimators to counteract their initialization at the origin. This helps in taking an adaptive learning rate method that can handle sparse gradients on noisy problems.

###The key difference between SGD and Adam can be summarized as follows:

`Learning Rate`: SGD maintains a constant learning rate for all weight updates and the learning rate does not change during training, whereas Adam adapts the learning rate for each weight.

## `Effects of Batch Size on Training`
### Learning Dynamics:

- Small Batches: Lead to more frequent updates, potentially causing faster convergence. However, they can also cause the training to be more unstable and less consistent.
- Large Batches: Provide a more accurate estimate of the gradient, but with less frequent updates. This can lead to more stable, but potentially slower, convergence.
### Computational Efficiency:

- GPU Utilization: Larger batches can utilize GPU resources more effectively as parallel processing is more efficient with larger amounts of data.
- Memory Constraints: However, large batches are limited by the memory capacity of the GPU. If a batch is too large to fit into memory, it can't be processed.
### Generalization:

Studies have shown that smaller batch sizes often lead to models that generalize better to new data. This is believed to be due to the noise in the estimation of the gradient, which can help the model escape local minima in the loss landscape.
Conversely, larger batches tend to converge to sharper minima, which might not generalize as well.
### Training Time:

While larger batches might process more data at once, this doesn't always mean faster training. The relationship between batch size and training speed is complex and depends on various factors, including the efficiency of parallel computation and the specific architecture of the neural network.
### Hyperparameter Interaction:

Batch size can interact with other hyperparameters such as learning rate and momentum. Often, when using larger batch sizes, adjustments to these other hyperparameters are necessary.

# Regularization
nn.Dropout(0.5) refers to a dropout layer where the probability (p) of each neuron being dropped ( temporarily removed from the network) during training is set to 0.5, or 50%. This means that during each training step, each neuron (or unit) has a 50% chance of being excluded from the computation.

### Dropout and Batch Normalization are two techniques used in neural networks:

- Dropout: Randomly "turns off" a proportion of neurons in the network during training. This prevents neurons from co-adapting too much, reduces overfitting, and leads to a more robust network. It's like training a network ensemble within a single model.

- Batch Normalization: Normalizes the inputs of each layer for each mini-batch. This stabilizes learning by reducing internal covariate shift (changes in the distribution of network activations), allows for higher learning rates, and reduces the sensitivity to weight initialization.

```python
def sse(y_true, y_pred):
    squared_errors = (y_true - y_pred) ** 2
    sse = np.sum(squared_errors)
    return sse
```

# Evaluation and analysis
## Question1:Research about each of the criterias (for measuring and evaluating the performance of models) below and explain how each is calculated:Accuracy ,Precision, RecallF1, Score

1. **Accuracy**: Accuracy measures the overall correctness of the model's predictions by comparing the number of correctly classified instances to the total number of instances. It is calculated using the following formula:

$$Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$  

where TP (True Positive) is the number of correctly predicted positive instances, TN (True Negative) is the number of correctly predicted negative instances, FP (False Positive) is the number of incorrectly predicted positive instances, and FN (False Negative) is the number of incorrectly predicted negative instances.

2. **Precision**: Precision focuses on the correctness of positive predictions made by the model. It measures the proportion of correctly predicted positive instances out of all the instances predicted as positive. Precision is calculated as:

$$Precision = \frac{TP}{TP + FP}$$  

Precision is useful when the cost of false positives is high, and it is important to minimize false positive predictions.

3. **Recall**: Recall, also known as sensitivity or true positive rate, measures the model's ability to identify positive instances correctly. It calculates the proportion of correctly predicted positive instances out of all actual positive instances. The formula for recall is:

 $$Recall = \frac{TP}{TP + FN}$$

Recall is particularly important when the cost of false negatives is high, and it is crucial to avoid missing positive instances.

4. **F1 Score**: The F1 score is a metric that combines precision and recall into a single value, providing an overall measure of a model's performance. It is the harmonic mean of precision and recall and is calculated as:

$$F1Score =2* \frac{Precision*Recall}{Precision + Recall}$$

The F1 score is useful when there is a trade-off between precision and recall, and we want to consider both measures simultaneously.
