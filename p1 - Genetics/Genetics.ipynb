# Part2: Production of the primary population 
in the "generate_coefficients" function I tried to generate the primary population.Due to that, I used "random" library to generate a row of random coefficients. and I repeated it to expand my population and have more rows.
here we recieve the coefficient range from the highest to the lowest, number of rows(chromosomes or population) , and degree of the polynomial.


```python
#row= population
def generate_coefficients(high_co, low_co , row , degree):
    coefs = []
    for i in range(0,row):
        coefs.append((random.sample(range(low_co,high_co+1), degree + 1)))
    for row in coefs:
        row.append(0)
    return coefs
```

# Part3: Implementation and specification of the fitness function

SSE(Sum of squared errors)is the sum of the squares of residuals (deviations predicted from actual empirical values of data). It is a measure of the discrepancy between the data and an estimation model, such as a linear regression. A small RSS indicates a tight fit of the model to the data. It is used as an optimality criterion in parameter selection and model selection.

$$ {\displaystyle \operatorname {SSE} =\sum _{i=1}^{n}(y_{i}-f(x_{i}))^{2}} $$

where yi is the ith value of the variable to be predicted, xi is the ith value of the explanatory variable, and f(x_{i}) is the predicted value of yi 

```python
def SumOfSqueredErrors  (points ,coefs):
    for row in coefs:
        sse = 0
        for point in points:
            x,y = point
            y_predict =0 
            for degree in range(len(row)):
                coef = row[degree]
                y_predict += coef*(x**degree)
            sse += (y - y_predict)**2
        row[-1] = sse
    return coefs
```

in selection step, the population is got as the input. then we sort and select 50% of the better population.(we sort it according to the SSE rate). Then we randomize them to get better results.

```python
def selection(population ):
    pop = sorted(population, key=itemgetter(-1))[0:int(len(population)*0.5)]
    if(len(pop)%2 != 0):
        pop.append(copy.deepcopy(pop[0]))
    random.shuffle(pop)
    return pop
```

# Part4: Implementing the Genetic Algorithm on the problem 

* Crossover:crossover, also called recombination, is a genetic operator used to combine the genetic information of two parents to generate new offspring. It is one way to stochastically generate new solutions from an existing population.

* Mutation: A mutation is a random change to a solution in a population of solutions. Mutations can be beneficial, harmful, or neutral to the solution's fitness. In artificial intelligence, mutations are often used to generate new solutions in the hope of finding a better solution.

In my crossover method, I randomly selected a point and moved all the genes after that point with the other parent.
 I tried different crossover but this one, gave me better results. In my case I gave "PC" as the input of my function which stands for the "Probability of crossover" and it changes the chromosomes differently.



```python
def crossover(population , points , Pc): #pc is a probability between 0 and 1
    childs = []
    childs = copy.deepcopy(population)
    degree = len(population[0]) - 1
    for i in range(0 , len(childs) , 2):
        for d in range(degree):
            pc = random.uniform(0, 1)
            if(pc > (100-Pc)/100):
                temp = childs[i][d]
                childs[i][d] = childs[i+1][d]
                childs[i+1][d] = temp
        childs[i][-1] = 0
        childs[i + 1][-1] = 0
    childs = SumOfSqueredErrors (points ,childs)
    result = childs + population
    return result
    
```

now, we do a mutation operation on the population. In order to generate new properties we added this function. So we will get better results. PM stands for "probability of mutation" and plays a key role in this algorithm.

```python
def mutation(population , low_co , high_co , Pm):
    degree = len(population[0]) - 1
    for i in population:
        for d in range(degree):
            pm = random.uniform(0, 1)
            if (pm > (100-Pm)/100):
                i[d] = random.sample(range(low_co,high_co+1), 1)[0]
        i[-1] = 0
    population = SumOfSqueredErrors (points ,population)
    return population
```

# Part5: Do the genetic algorithm on the problem

at the end, we consider all the previous steps and create a GA function to implement the genetic algorithm. this function iterates all the above steps in the number of epochs.
- I made my function more efficient in this way:
in the number of iterations if  (SSE== 0) -> all the proccess will be stopped and the correct coefficients will be returned.
(SSE=0 ) means we have reached the accurate answer.
this decreased my runtime alot and gave me better answers.

```python
def GA(high_co, low_co , row , degree , points , epochs , Pm , Pc):
    population = generate_coefficients(high_co, low_co , row , degree)
    for i in range(epochs):
        population = SumOfSqueredErrors(points ,population)
        population = selection(population)
        population = crossover(population , points , Pc)
        population = mutation(population , low_co , high_co , Pm)
        population = sorted(population, key=itemgetter(-1))
        x = population[0][-1]
        print("iteration " + str(i) + "th  SSE = " + str(x))
        if(x ==0 ): #early drop
            return sorted(population, key=itemgetter(-1))[0][0:-1]
        random.shuffle(population)
  
    return sorted(population, key=itemgetter(-1))[0]
```
